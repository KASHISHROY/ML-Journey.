```python
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def bark(self):
        print(f"{self.name} says woof")
dog1=Dog("Buddy",3)
dog2=Dog("Leo",9)
dog1.bark()
dog2.bark()
```

```python
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount}deposited,The new Balance is {self.balance}")
    def withdraw(self,amount):
        if(self.balance<amount):
            print("Cant withdrwa this amount")
        else:
            self.balance-=amount
            print(f"{amount}withdrawn,The new Balance is {self.balance}")
    def getbalance(self):
        return self.balance
acnt1=BankAccount("KASHISH",10000)
print(acnt1.getbalance())
acnt1.deposit(5000)
acnt1.withdraw(1000000)
```
