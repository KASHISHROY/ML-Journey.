```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
##combine the two regions into one dataset
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
##header=1 as we are not considering the first row as header where the region name is written
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
##while mergin there was one row which just had the region detail that also got merged hence all the values are none there
```

```python
#The dataset is converted into two sets based on region from 122th index ,we can make a new column based on the region.
#1:"Whatever region"->label 0
#2:"Whatever othere region"->label 1
#add a new column with region
```

```python
#loc till 122 have label=1 and afer 122 label 1 for region column
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
#region no need of saving in float,so we can convert it into int
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
##remove the null value,only one actual record with null value we can drop it.
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
#this row is basically the repeated header again which we were using for the region 2 before merging.
```

```python
#remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##fix spaces in column names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
##all are object but if you see the dataset then all are almost newmeric value,except the classes column which is categorical
```

```python
##change the required columns as integer data type
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
##changing the other columns to float datatype
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
