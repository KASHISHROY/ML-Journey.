```python
data=np.array([1,2,3,4,5,6])
print(np.mean(data))
print(np.median(data))
print(np.var(data))
print(np.std(data))
```

```python
data=np.array([1,2,3,4,5])
mean=np.mean(data)
std_dev=np.std(data)
n_data=(data-mean)/std_dev
print(n_data)
print(np.mean(n_data))
print(np.std(n_data))
```
