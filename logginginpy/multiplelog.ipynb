```python
import logging
```

```python
logger1=logging.getLogger("module 1")
logger1.setLevel(logging.DEBUG)

logger2=logging.getLogger("module 2")
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%y-%m-%d %H:%M:%S'
)
```

```python
logger1.debug("This is a debug messages from module 1")
logger2.warning("This is a warning message from module 2")
logger2.critical("This is a critical message from module 2")
```
