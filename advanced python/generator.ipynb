```python
def square(n):
    for i in range(n):
        yield i**2
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
gen=my_generator()
gen

for val in gen:
    print(val)
```

```python
def read_large_files(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
file_path="example.txt"
for line in read_large_files(file_path):
    print(line.strip())
```
